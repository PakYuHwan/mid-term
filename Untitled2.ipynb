```r

KOSPI_T<-t(KOSPI)[2:308,]

colnames(KOSPI_T)<-c('date','KOSPI')
KOSPI_T<-as.data.frame(KOSPI_T)
KOSPI_T$KOSPI<-as.numeric(KOSPI_T$KOSPI)
KOSPI_T$date<-gsub(". ","-",KOSPI_T$date)

head(KOSPI_T)
```

로그 변환도 취해봤지만 딱히 의미있어보이진 않는다.

문득 든 생각이 코로나는 바이러스고, KOSPI는 주식이라는 점이다.
다시 말하면 코로나 확진자수는 어제가 100명이었어도 다음 날 1000명의 확진자가 발생할 순 있지만 KOSPI는 절대 그럴 수 없다는 점
(주식은 다음날 아무리 뛰어올라봐야 30%라는 상한선이 있기 때문에)

따라서 거리두기 기준+최근 4월 확진자 평균보다는 낮은 600명을 기준으로 데이터를 잘라서 분석해봤다.

```r
joined_600<-filter(joined_2,new_cases<600)
```

```r
plot(joined_600$new_cases,joined_600$KOSPI)
plot(joined_600$new_cases,log(joined_600$KOSPI))
plot((joined_600$new_cases)^0.5,(joined_600$KOSPI)^0.5)
```

```r
c<-lm(KOSPI~new_cases,data=joined_600)
summary(c)
```

와! 행복하다 통계학과 거리두기를 한지 꽤나 오래되어 돌릴 수 있는 거라곤 선형회귀 분석정도인데 나름 의미가 있는 수치들이 나왔다.

-R^2 설명력에 의하면 대략 70%정도의 설명령을 갖고

-확진자수 증가에 따라 코스피 지수도 올라간다고 봐도 무방한 수치가 나왔다.(회귀계수=0에 대한 귀무가설 기각)



그래도 데이터를 아예 버리기보단 확진자수에 대한 영향력을 줄이면 어떨까 싶어
 0.95승을 하여 영향력을 줄여서 분석을 다시 해봤다.

```r
1000^0.95
#600보단 높긴하지만 그래도 600명대에 비하면 높은 수치라 생각되기에 0.95승으로 했음.
```

* 여기서 2번째로 급진적으로 코로나가 확산되는 상황 즈음엔 여기저기서 코스피 지수가 이전보다 더 높아졌다는 소식이 들렸을 것이다.
